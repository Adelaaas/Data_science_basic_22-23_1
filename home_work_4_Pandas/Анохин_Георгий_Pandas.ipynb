```python
def get_date(row, q):
  if row['Дата выпуска'] is np.nan:
    return np.nan
  return str(row['Дата выпуска']).split("-")[q]


studs_info['Год выпуска'] = studs_info.apply(lambda row: get_date(row, 0), axis=1)
studs_info['Месяц выпуска'] = studs_info.apply(lambda row: get_date(row, 1), axis=1)
studs_info['День выпуска'] = studs_info.apply(lambda row: get_date(row, 2), axis=1)


studs_info.head()
```

```python
ar = studs_info['Категория обучения'].unique()
print(ar)
print(f"Кол-во уникальных значений в 'Категория обучения' >>> {len(ar)}")
print("Кол-во уникальных (не нулевых) значений в 'Категория обучения' >>> ", end="")

for el in ar:
  if el is np.nan:
    print(len(ar) - 1)
    break
else:
  print(len(ar))
```

```python
ar = np.append(ar, [None])
m = {ar[i]: i for i in range(len(ar))}

studs_info['Категория обучения'] = studs_info.apply(lambda row: m[row['Категория обучения']], axis=1)
print(studs_info['Категория обучения'].unique())
(studs_info.head())
```

```python
studs_info.drop("Дата выпуска", axis=1, inplace=True)
studs_info
```
