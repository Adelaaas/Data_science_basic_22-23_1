3. Построить boxplot для всех числовых переменных. График должен быть один, используя разные subplot() для каждой переменной (т.е. одна фигура и несколько сабплотов).

**Датасет для задания называется train_var_3.csv**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_3.csv')
df.head()
df.count()
```

1.

```python
fig, ax = plt.subplots(figsize=(7, 7))
mono_font = {'fontname':'monospace'}
x_ax, y_ax  = df['MinTemp'], df['MaxTemp']
plt.title("Диаграмма разброса температур", **mono_font)
plt.xlabel("Min temp", **mono_font)
plt.ylabel("Max temp", **mono_font)
ax.scatter(x_ax, y_ax, s = 0.2)
plt.show()
```

2.

```python
null_count = df.isna().sum()
mono_font = {'fontname':'monospace'}
plt.plot(null_count.values, null_count.keys())
plt.xlabel("Кол-во пропусков", fontsize=14)  
plt.ylabel("Столбцы", fontsize=14)
plt.title("Диаграмма разброса пропусков", **mono_font)
plt.show()
```

Больше всего пропусков в "Evaporation", "Sunshine", "Cloud9am", "Cloud3pm"

3.

```python
columns = df.select_dtypes(include='number')
fig, axes = plt.subplots(1, len(list(columns.columns)), figsize=(len(list(columns.columns)) * 10, 8))  
for i, k in enumerate(columns.columns):
    sns.boxplot(columns[k], ax=axes[i], width=0.3)
plt.show()
```

4.

```python
sns.barplot(data = df, y = 'Location', x = 'RainTomorrow')
plt.xlabel("Количество дождливых дней", fontsize=14)  
plt.ylabel("Город", fontsize=14)
plt.title("Количество дождливых дней в городах")
plt.show()
```

5.

```python
plt.figure(figsize=(15,10))
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu_r', center=0, annot=True)
plt.title('Матрица корреляции переменных')
plt.show()
```

6.

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df.select_dtypes(include='number'))
plt.show()
```
