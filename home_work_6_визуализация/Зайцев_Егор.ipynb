```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_2.csv')
df.head()
```

```python
fig, ax = plt.subplots()

x = df['age']
y = df['balance']

plt.xlabel('age')
plt.ylabel('balance')

ax.scatter(x, y)

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
```

```python
fig, axes = plt.subplots(4, 2, figsize=(40, 20))  

columns = ['age', 'balance',  'day', 'duration', 'campaign', 'pdays', 'previous']

print(columns)

indcol = 0
for y in range(4):
  for x in range(2):
    col = columns[indcol]
    sns.boxplot(df[col], ax=axes[y][x], width=0.4)
    indcol += 1
     
```

```python
plt.figure(figsize=(10,10))

deposit = df['deposit']
yes = len(deposit[deposit == 'yes'])
no = len(deposit[deposit == 'no'])

arr = pd.Series([yes, no])

sns.barplot(df['deposit'], arr)

plt.show()
     
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
     
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.pairplot(df, kind="reg", hue="deposit")

plt.show()
     
```
