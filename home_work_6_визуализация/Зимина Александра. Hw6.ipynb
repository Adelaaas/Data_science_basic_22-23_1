```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_2.csv')
df.head()
```

#№1

```python
x = df["age"]
y = df["balance"]

fig, ax = plt.subplots(figsize=(12, 10))

ax.scatter(x, y, c = 'r')

plt.title("Диаграмма разброса точек age и balance", fontsize=22)
plt.xlabel("age", fontsize=12) 
plt.ylabel("balance", fontsize=12) 
plt.show()
```

Из графика видно что точки, определяемые age и balance, разбросаны не очень сильно.

#№2

```python
plt.figure(figsize=(12,10))

cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Пропущенные значения'})

plt.title("Пропущенные значения", fontsize=22)
```

Из графика, видно, что в данном датасете пропусков нет.

#№3

```python
fig, axes = plt.subplots(4, 2, figsize=(40, 20))  

sns.boxplot(df['age'], ax=axes[0][0], width=0.3)
sns.boxplot(df['balance'], ax=axes[0][1], width=0.3)
sns.boxplot(df['day'], ax=axes[1][0], width=0.3)
sns.boxplot(df['duration'], ax=axes[1][1], width=0.3)
sns.boxplot(df['campaign'], ax=axes[2][0], width=0.3)
sns.boxplot(df['pdays'], ax=axes[2][1], width=0.3)
sns.boxplot(df['previous'], ax=axes[3][0], width=0.3)
```

```python
plt.figure(figsize=(12,10))

df[['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']].plot(kind='box')

plt.title("Boxplot для всех числовых переменных", fontsize=22)
```

#№4

```python
plt.figure(figsize=(10,10))

y = [df[df["deposit"] == 'yes']["deposit"].count(), df[df["deposit"] == 'no']["deposit"].count()]
y = pd.Series(y)

sns.barplot(df["deposit"], y)

plt.xlabel("Решение", fontsize=12)  
plt.ylabel("Количество решений", fontsize=12)
plt.title("Выдали депозит?", fontsize=22)

plt.show()
```

Из таблицы видно, что данные по решению выдачи депозита распределены равномерно, но yes встречается немного чаще, чем no.

#№5

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Матрица корреляции', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Из матрицы видно, что больше всего между собой коррелируют pdays и previous.

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.pairplot(df, kind="reg", hue="deposit")

plt.show()
```
