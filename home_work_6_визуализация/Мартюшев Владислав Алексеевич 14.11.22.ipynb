**Датасет для задания называется train_var_3.csv**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_3.csv')
```

1

```python
fig, ax = plt.subplots()
x = df['MinTemp']
y = df['MaxTemp']
plt.xlabel("MinTemp", fontsize=14)  
plt.ylabel("MaxTemp", fontsize=14)
plt.title("Разброс минимальных и максимальных суточных температур")
plt.grid()
plt.minorticks_on()
ax.scatter(x, y, c = 'deeppink', s = 0.5)
plt.show()
```

2

```python
plt.figure(figsize=(15,7))
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(df.isna().transpose(), cmap=cmap, cbar_kws={'label': 'Missing Data'})
plt.xlabel("Индексы", fontsize=14)  
plt.ylabel("Критерии", fontsize=14)
plt.title("Количество пропусков в данных")
plt.grid()
plt.show()
```

3

```python
fig, axes = plt.subplots(1, 18)
plt.figure(figsize=(3,6), dpi= 80)

sns.boxplot(df['MinTemp'], ax=axes[0], width=0.3)
sns.boxplot(df['MaxTemp'], ax=axes[1], width=0.3)
sns.boxplot(df['Rainfall'], ax=axes[2], width=0.3)
sns.boxplot(df['Evaporation'], ax=axes[3], width=0.3)
sns.boxplot(df['Sunshine'], ax=axes[4], width=0.3)
sns.boxplot(df['WindGustSpeed'], ax=axes[5], width=0.3)
sns.boxplot(df['WindSpeed9am'], ax=axes[6], width=0.3)
sns.boxplot(df['WindSpeed3pm'], ax=axes[7], width=0.3)
sns.boxplot(df['Humidity9am'], ax=axes[8], width=0.3)
sns.boxplot(df['Humidity3pm'], ax=axes[9], width=0.3)
sns.boxplot(df['Pressure9am'], ax=axes[10], width=0.3)
sns.boxplot(df['Pressure3pm'], ax=axes[11], width=0.3)
sns.boxplot(df['Cloud9am'], ax=axes[12], width=0.3)
sns.boxplot(df['Cloud3pm'], ax=axes[13], width=0.3)
sns.boxplot(df['Temp9am'], ax=axes[14], width=0.3)
sns.boxplot(df['Temp3pm'], ax=axes[15], width=0.3)
sns.boxplot(df['RainToday'], ax=axes[16], width=0.3)
sns.boxplot(df['RainTomorrow'], ax=axes[17], width=0.3)
plt.show()
```

4

```python
sns.barplot(data = df, x = 'Location', y = 'RainTomorrow')
plt.xlabel("Город", fontsize=14)  
plt.ylabel("Количество дождливых дней", fontsize=14)
plt.title("Количество дождливых дней в городах")
plt.grid()
plt.show()
```

5

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Корреляции в характеристиках', fontsize=20)
plt.show()
```
