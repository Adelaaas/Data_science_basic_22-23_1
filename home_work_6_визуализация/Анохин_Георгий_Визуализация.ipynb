```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train.csv')
df.head()
```

```python
# 1
fig, ax = plt.subplots()
plt.xlabel("Vmag")        
plt.ylabel("e_Plx")

ax.scatter(df['Vmag'], df['e_Plx'], c = 'black')


plt.show()
```

```python
# 2
plt.figure()
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
```

```python
# 3
fig, axes = plt.subplots(2, 3, figsize=(12,12))
c = 0
for name in df.keys():
  if name == 'SpType':
    continue
  ax = axes[c//3][c%3]
  ax.set_ylabel(name, fontsize=14)
  sns.boxplot(df[name], ax=ax, width=0.25)
  c += 1
```

```python
# 4
data = df['TargetClass'].value_counts()
plt.xlabel('Класс звезды')
plt.ylabel('Кол-во')
sns.barplot(x=data.keys(), y=data)
```

```python
# 5

plt.figure(figsize=(7, 7), dpi=130)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True, )
plt.title('Корреляции')
plt.show()
```

```python
# 6

plt.figure(figsize=(10,8), dpi= 100)
sns.pairplot(df)
plt.show()
```
