```python
x = df.loc[:,['BALANCE','PURCHASES','PAYMENTS']]
SSE = []
for k in range(1,10):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(x)
    SSE.append(kmeans.inertia_)
 
plt.plot(range(1,10), SSE, marker='s');
plt.xlabel('Число кластеров')
plt.ylabel('SSE');
plt.title('Метод локтя для малого чилса кластеров')
plt.show()


     
     
```

```python
x = df.loc[:,['BALANCE','PURCHASES','PAYMENTS']]
SSE = []
for k in range(1,50):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(x)
    SSE.append(kmeans.inertia_)
 
plt.plot(range(1,50), SSE, marker='s');
plt.xlabel('Число кластеров')
plt.ylabel('SSE');
plt.title('Метод локтя для большого чилса кластеров')
plt.show()

```
