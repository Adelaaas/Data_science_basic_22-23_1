```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_2.csv')
df.head()
```

```python
#Задание 1

fig, ax = plt.subplots()

ax.scatter(df.age, df.balance, c = 'blue')
plt.xlabel("age")
plt.ylabel("balance")

plt.show()
```

```python
#Задание 2

plt.figure(figsize=(15,7))

cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
```

```python
#Задание 3

fig1, axes = plt.subplots(7, 1, figsize=(10,25))

sns.boxplot(df['age'], ax=axes[0], width=0.3)
sns.boxplot(df['balance'], ax=axes[1], width=0.3)
sns.boxplot(df['day'], ax=axes[2], width=0.3)
sns.boxplot(df['duration'], ax=axes[3], width=0.3)
sns.boxplot(df['campaign'], ax=axes[4], width=0.3)
sns.boxplot(df['pdays'], ax=axes[5], width=0.3)
sns.boxplot(df['previous'], ax=axes[6], width=0.3)
```

```python
#Задание 4

plt.hist(df.deposit)
```

```python
#Задание 5

plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram of DataFrame', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
#Задание 6

plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df)
plt.show()
```
