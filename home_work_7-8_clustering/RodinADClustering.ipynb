```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/home_work_7-8_clustering/Credit%20Card%20Dataset.csv', sep=',')
df
```

```python
plt.figure(figsize=(10,6))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=2
)
```

```python
plt.figure(figsize=(10, 10), dpi=80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True, )
plt.title('ÐšÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¸')
plt.show()
```

```python
df = df.drop('CUST_ID', axis = 1)
df = df.dropna(axis=0, how='any')
SSE = []
for i in range(1,60):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df)
  SSE.append(kmeans.inertia_)
plt.plot(range(1,60), SSE, marker='s');
plt.title('ÐÐ½Ð°Ð»Ð¸Ð· Ð¼ÐµÑ‚Ð¾Ð´Ð¾Ð¼ Ð»Ð¾ÐºÑ‚Ñ')
plt.xlabel('ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð²')
plt.ylabel('Ð¡Ñ€ÐµÐ´Ð½ÐµÐµ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ (SSE)')
plt.show()
```

```python
x = list( df['BALANCE'] )
y = list( df['BALANCE_FREQUENCY'] )
fig, ax = plt.subplots()
ax.scatter(x, y, c = 'lightgreen')
plt.title('Ð”Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸Ð·Ð°Ñ†Ð¸Ð¸ (Ð¿Ð¾ Ð´Ð²ÑƒÐ¼ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ð¼)')
plt.xlabel('BALANCE')
plt.ylabel('BALANCE_FREQUENCY')
plt.show()
```

```python
SSE = []
df2 = df[['BALANCE', 'BALANCE_FREQUENCY']]
for i in range(1,60):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df2)
  SSE.append(kmeans.inertia_)
plt.plot(range(1,60), SSE, marker='s');
plt.title('ÐœÐµÑ‚Ð¾Ð´ Ð»Ð¾ÐºÑ‚Ñ (Ð´Ð»Ñ Ð´Ð²ÑƒÑ… Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ñ…)')
plt.xlabel('ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¾Ð²')
plt.ylabel('Ð¡Ñ€ÐµÐ´Ð½ÐµÐµ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ (SSE)')
plt.show()
```

```python
n_clust_df2 = 6
kmeans = KMeans(n_clusters=n_clust_df2)
kmeans.fit(df2)
Y_pred = kmeans.labels_
df2['cluster'] = Y_pred
c = ['m','y','k','g','r','c','b']
fig, ax = plt.subplots()
for i in range(n_clust_df2):
    ax.scatter(df2[df2['cluster'] == i]['BALANCE'], df2[df2['cluster']==i]['BALANCE_FREQUENCY'], c = c[i])
plt.title('Ð”Ð¾ ÐºÐ»Ð°ÑÑ‚ÐµÑ€Ð¸Ð·Ð°Ñ†Ð¸Ð¸ (Ð¿Ð¾ Ð´Ð²ÑƒÐ¼ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ð¼)')
plt.xlabel('BALANCE')
plt.ylabel('BALANCE_FREQUENCY')
plt.show()
```
