```python
def date(a, b):
  if a['Дата выпуска'] is np.nan:
    return np.nan
  return str(a['Дата выпуска']).split("-")[b]


studs_info['Год выпуска'] = studs_info.apply(lambda a: date(a, 0), axis=1)
studs_info['Месяц выпуска'] = studs_info.apply(lambda a: date(a, 1), axis=1)
studs_info['День выпуска'] = studs_info.apply(lambda a: date(a, 2), axis=1)


studs_info.head()
```

```python
katOb = studs_info['Категория обучения'].unique()
print(katOb)
print(f"Кол-во уникальных значений в 'Категория обучения' = {len(katOb)}")
print("Кол-во уникальных (не нулевых) значений в 'Категория обучения' = ", end="")

for i in katOb:
  if i is np.nan:
    print(len(katOb) - 1)
    break
else:
  print(len(katOb))
```

```python
array = np.append(katOb, [None])
a = {array[i]: i for i in range(len(array))}

studs_info['Категория обучения'] = studs_info.apply(lambda b: a[b['Категория обучения']], axis=1)
print(studs_info['Категория обучения'].unique())
(studs_info.head())
```

```python
studs_info.drop("Дата выпуска", axis=1, inplace=True)
studs_info
```
