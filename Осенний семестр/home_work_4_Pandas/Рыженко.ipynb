```python
import pandas
import numpy

url_x_train_csv = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_4_Pandas/ONTI%20Students%20performance/X_train.csv'
url_y_train_csv = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_4_Pandas/ONTI%20Students%20performance/y_train.csv'
url_studs_info_csv = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_4_Pandas/ONTI%20Students%20performance/studs_info.csv'
url_zachety_csv = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_4_Pandas/ONTI%20Students%20performance/zachety.csv'

x_train_csv = pandas.read_csv(url_x_train_csv)
y_train_csv = pandas.read_csv(url_y_train_csv)
studs_info_csv = pandas.read_csv(url_studs_info_csv)
zachety_csv = pandas.read_csv(url_zachety_csv)
```

```python
# 1

print(x_train_csv.shape, y_train_csv.shape, studs_info_csv.shape, zachety_csv.shape)
```

```python
# 2

train = pandas.concat([x_train_csv, y_train_csv],axis=1)
train
```

```python
# 4

print(x_train_csv['АТТЕСТАЦИЯ'].unique())
print(x_train_csv['АТТЕСТАЦИЯ'].value_counts())
```

```python
# 5

train.columns = [i.lower() for i in train.columns]
train
```

```python
# 6

dc = {'Код направления':[],'Название направления': []}
for i in range(train.shape[0]):
    kod, nazv = map(str, str(train.iloc[i][2]).split(maxsplit=1))
    dc['Код направления'].append(kod)
    dc['Название направления'].append(nazv)

df = pandas.DataFrame(dc)
train = pandas.concat([train,df],axis=1)
train.drop(['направление'],axis=1,inplace=True)

train
```

```python
# 7

std_id_male = studs_info_csv[studs_info_csv['Пол'] == 'М']['STD_ID'].values
train_male_3year = train[train['std_id'].isin(std_id_male) == True][train['курс'] == 3]
train_male_3year
```

```python
# 8

d = {'Пол': []}
for i in range(studs_info_csv.shape[0]):
  d['Пол'].append(0  if (studs_info_csv.loc[i]['Пол']=='М') else  1)

studs_info_csv_intGender = pandas.DataFrame(d)

studs_info_csv.drop(['Пол'], axis=1, inplace=True)
studs_info_csv = pandas.concat([studs_info_csv,studs_info_csv_intGender],axis=1)
studs_info_csv
# d = {'Пол':[]}
# d['Пол'].append( 0  if studs_info_csv.loc[0]['Пол']=='М' else  1)
# d
# studs_info_csv.shape[0]/2
# studs_info_csv
```
