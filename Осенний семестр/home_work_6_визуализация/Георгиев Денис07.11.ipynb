**Датасет для задания называется train_var1.csv**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train.csv')
df.head()
```

```python
#1
plt.scatter(df['Vmag'], df['e_Plx'])
plt.xlabel("Vmag")
plt.ylabel("e_Plx")
plt.show()
```

```python
#2
sns.displot(data=df.isna().melt(value_name="missing"))
plt.show()
```

```python
#3
fig, axes = plt.subplots(2, 3, figsize=(12,12))
st = 0
for key in df.keys():
	if df[key].dtype == np.int64 or df[key].dtype == np.float64:
		axes[st//3][st%3].boxplot(df[key])
		axes[st//3][st%3].set_title(key, fontsize=14) 
		st += 1
plt.show()
```

```python
#4
d = df['TargetClass'].value_counts()
sns.barplot(x=d.keys(), y=d)
plt.show()
```

```python
#5
plt.figure(figsize=(10, 10), dpi=100)
sns.heatmap(
	df.corr(), 
	xticklabels=df.corr().columns, 
	yticklabels=df.corr().columns, 
	cmap='RdYlGn', 
	center=0, 
	annot=True, 
)
plt.show()
```

```python
#6
plt.figure(figsize=(10,8))
sns.pairplot(df)
plt.show()
```
