**Датасет для задания называется train_var_3.csv**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/class_work_6/train_var_3.csv')
df.head()
```

```python
# 1. Постройте Диаграмму разброса (точечный график) от двух переменных: MinTemp и MaxTemp.

fig, ax = plt.subplots()
x = df['MinTemp']
y = df['MaxTemp']
plt.title("Диаграмма разброса")
plt.xlabel("Минимальная температура")
plt.ylabel("Максимальная температура")
ax.scatter(x,y)
plt.show()
```

```python
# 2. Построить график анализа пропущенных значений.

plt.figure(figsize=(15,7))
plt.title("График анализа пропущенных значений")
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
plt.xlabel("Индексы")
plt.ylabel("Признаки")
plt.show()
```

```python
# 3. Построить boxplot для всех числовых переменных. График должен быть один, используя разные subplot() для каждой переменной
#    (т.е. одна фигура и несколько сабплотов).

fig, axes = plt.subplots(1, 22)
plt.figure(figsize=(3,6), dpi= 80)

# c = 0
# for i in df.keys():
    #     sns.boxplot(df[i], ax=axes[c])
    #     c+=1
    
    # for i in range(len(df.keys())):
    #     sns.boxplot(df[str(df.keys()[i])], ax=axes[i], width=0.3)
    
sns.boxplot(df['MinTemp'], ax=axes[0], width=0.3)
sns.boxplot(df['MaxTemp'], ax=axes[1], width=0.3)
sns.boxplot(df['Rainfall'], ax=axes[2], width=0.3)
sns.boxplot(df['Evaporation'], ax=axes[3], width=0.3)
sns.boxplot(df['Sunshine'], ax=axes[4], width=0.3)
sns.boxplot(df['WindGustSpeed'], ax=axes[5], width=0.3)
sns.boxplot(df['WindSpeed9am'], ax=axes[6], width=0.3)
sns.boxplot(df['WindSpeed3pm'], ax=axes[7], width=0.3)
sns.boxplot(df['Humidity9am'], ax=axes[8], width=0.3)
sns.boxplot(df['Humidity3pm'], ax=axes[9], width=0.3)
sns.boxplot(df['Pressure9am'], ax=axes[10], width=0.3)
sns.boxplot(df['Pressure3pm'], ax=axes[11], width=0.3)
sns.boxplot(df['Cloud9am'], ax=axes[12], width=0.3)
sns.boxplot(df['Cloud3pm'], ax=axes[13], width=0.3)
sns.boxplot(df['Temp9am'], ax=axes[14], width=0.3)
sns.boxplot(df['Temp3pm'], ax=axes[15], width=0.3)
sns.boxplot(df['RainToday'], ax=axes[16], width=0.3)
sns.boxplot(df['RainTomorrow'], ax=axes[17], width=0.3)

plt.show()
```

```python
# 4. Построить barplot по переменной RainTomorrow.

plt.title("Количество дождей")
sns.barplot(data = df, x = 'Location', y = 'RainTomorrow')
plt.xlabel("Города")
plt.ylabel("Количество дождей завтра")
plt.show()
```

```python
# 5. Построить матрицу корреляции для всех переменных.

plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('матрицу корреляции для всех переменных', fontsize=20)
plt.show()
```

```python
# 6. Построить графики парных зависимостей между всеми числовыми переменными.

plt.figure(figsize=(36,9), dpi= 80)
sns.pairplot(df)
plt.show()
```
