```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_22-23_1/main/home_work_7-8_clustering/Credit%20Card%20Dataset.csv', sep=',')
df
```

```python
plt.figure(figsize=(10,6))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=2
)
```

```python
plt.figure(figsize=(10, 10), dpi=80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True, )
plt.title('Корреляции')
plt.show()
```

```python
df = df.drop('CUST_ID', axis = 1)
df = df.dropna(axis=0, how='any')
SSE = []
for i in range(1,60):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df)
  SSE.append(kmeans.inertia_)
plt.plot(range(1,60), SSE, marker='s');
plt.title('Анализ методом локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('Среднее расстояние (SSE)')
plt.show()
```

```python
x = list( df['BALANCE'] )
y = list( df['BALANCE_FREQUENCY'] )
fig, ax = plt.subplots()
ax.scatter(x, y, c = 'lightgreen')
plt.title('До кластеризации (по двум переменным)')
plt.xlabel('BALANCE')
plt.ylabel('BALANCE_FREQUENCY')
plt.show()
```

```python
SSE = []
df2 = df[['BALANCE', 'BALANCE_FREQUENCY']]
for i in range(1,60):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(df2)
  SSE.append(kmeans.inertia_)
plt.plot(range(1,60), SSE, marker='s');
plt.title('Метод локтя (для двух переменных)')
plt.xlabel('Количество кластеров')
plt.ylabel('Среднее расстояние (SSE)')
plt.show()
```

```python
n_clust_df2 = 6
kmeans = KMeans(n_clusters=n_clust_df2)
kmeans.fit(df2)
Y_pred = kmeans.labels_
df2['cluster'] = Y_pred
c = ['m','y','k','g','r','c','b']
fig, ax = plt.subplots()
for i in range(n_clust_df2):
    ax.scatter(df2[df2['cluster'] == i]['BALANCE'], df2[df2['cluster']==i]['BALANCE_FREQUENCY'], c = c[i])
plt.title('До кластеризации (по двум переменным)')
plt.xlabel('BALANCE')
plt.ylabel('BALANCE_FREQUENCY')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
fig = plt.figure(figsize=(10, 10))
Z = linkage(df, 'ward')
dn = dendrogram(Z)
plt.title('Дендрограмма Ward Linkage')
plt.show()
```

```python
Z = linkage(df, method='average', metric='euclidean')
label = fcluster(Z, 20000, criterion='distance')
df['cluster'] = label
print(f'Кластеров: {len(np.unique(label))}')
df
```
