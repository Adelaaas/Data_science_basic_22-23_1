```python
import matplotlib.pyplot as plt

x = [11.5, 9.5, 11, 13.5, 17, 15.3, 14.5, 3.7, 5.4, 7.5]
y = [16.5, 15.5, 14.5, 7.5, 6.8, 5.6, 5, 7, 4.7, 4.3]

plt.figure(figsize=(12, 8), dpi=80)
plt.scatter(x,y, s=100)

for i, txt in enumerate(zip(x, y)):
    plt.annotate(txt, (x[i], y[i]))
```

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd

mas = pd.DataFrame({'x': x,'y': y})
kmeans = KMeans(n_clusters=3)
kmeans.fit(mas)
Y_pred = kmeans.labels_
Y_pred
```

```python
mas['cluster'] = Y_pred
plt.plot(mas[mas['cluster']==0]['x'], mas[mas['cluster']==0]['y'], 'bo', label='class1')
plt.plot(mas[mas['cluster']==1]['x'], mas[mas['cluster']==1]['y'], 'go', label='class2')
plt.plot(mas[mas['cluster']==2]['x'], mas[mas['cluster']==2]['y'], 'ro', label='class3')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=0)
plt.show()

```

```python
x = mas.loc[:,['x','y']]
SSE = []
for k in range(1,9):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(x)
    SSE.append(kmeans.inertia_)
plt.plot(range(1,9), SSE, marker='s');
plt.xlabel('k')
plt.ylabel('SSE');
plt.show()
```
